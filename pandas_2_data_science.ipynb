# merging and concat

# merging

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,4],"c":[19,15,10,19]})
pd.merge(var1,var2,on="A")# on-- jo same data hai ek baar likha jaye



```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"c":[19,15,10,19]})# mai yaha 5 likh diya hu jaha tk same rahega vahi tk nikalega
pd.merge(var1,var2,on="A")# on-- jo same data hai ek baar likha jaye

```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"c":[19,15,10,19]})
pd.merge(var1,var2,how="left")# on-- jo same data hai ek baar likha jaye

# left ka data vah data hai jo missing value ko nan karta
```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"c":[19,15,10,19]})
pd.merge(var1,var2,how="right")# on-- jo same data hai ek baar likha jaye

# right ka data vah data hai jo missing value ko nan karta
```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"c":[19,15,10,19]})
pd.merge(var1,var2,how="outer")# on-- jo same data hai ek baar likha jaye

# outer data dono defalt value dega
```

# Group by function

# groupby

```python
import pandas as pd

data =pd.DataFrame({
    "City": ["Lucknow", "Lucknow", "Kanpur", "Kanpur", "Delhi"],
    "Sales": [200, 250, 300, 150, 400]
})

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# Group by City and find total sales
result = df.groupby("City")["Sales"].sum()

print("\nTotal Sales in each City:")
print(result)
```

# pivot

```python
# pivot jo value chahiye vahi milega
import pandas as pd

data = {
    "Name": ["Sonu", "Sonu", "Riya", "Riya"],
    "Subject": ["Math", "Science", "Math", "Science"],
    "Marks": [85, 90, 80, 95]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Using pivot
pivot_df = df.pivot(index="Name", columns="Subject", values="Marks")

print("\nAfter Pivot:")
print(pivot_df)
```

# pivot()

```python
import pandas as pd

df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Mon', 'Tue'],
    'eng': ['A', 'A', 'B', 'B'],
    'score': [90, 95, 85, 80]
})

pivoted = df.pivot(index='day', columns='eng', values='score')
print(pivoted)
```

```python
import pandas as pd

df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Mon', 'Tue'],
    'eng': ['A', 'A', 'B', 'B'],
    'score': [90, 95, 85, 80]
})

pivoted = df.pivot(index='day', columns='eng')
print(pivoted)
```

```python
import pandas as pd

df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Mon', 'Tue', 'Mon'],
    'eng': ['A', 'A', 'B', 'B', 'A'],
    'score': [90, 95, 85, 80, 88]
})

pivoted = df.pivot_table(index='day', columns='eng', values='score', aggfunc='mean', fill_value=0)
print(pivoted)
```

```python
import pandas as pd

df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Mon', 'Tue', 'Mon'],
    'eng': ['A', 'A', 'B', 'B', 'A'],
    'score': [90, 95, 85, 80, 88]
})

pivoted = df.pivot_table(
    index='day',
    columns='eng',
    values='score',
    aggfunc='sum',
    margins=True 
)

print(pivoted)
```
