```python
pd.merge(df1,df2, how = "left")


# | id | name  |
# | -- | ----- |
# | 1  | Ram   |
# | 2  | Shyam |
# | 3  | Mohan |
# | id | name  | marks |
# | -- | ----- | ----- |
# | 1  | Ram   | 80    |
# | 2  | Shyam | NaN   |
# | 3  | Mohan | 90    |


# | id | marks |
# | -- | ----- |
# | 1  | 80    |
# | 3  | 90    |

# | id | name  | marks |
# | -- | ----- | ----- |
# | 1  | Ram   | 80    |
# | 2  | Shyam | NaN   |
# | 3  | Mohan | 90    |


```

```python
pd.merge(df1,df2, how = "right")#
#| id | marks |
# | -- | ----- |
# | 1  | 80    |
# | 3  | 90    |
# | 4  | 70    |


# | id | name  | marks |
# | -- | ----- | ----- |
# | 1  | Ram   | 80    |
# | 3  | Mohan | 90    |
# | 4  | NaN   | 70    |


# # IGHT ka seedha matlab 🧠

# # Right table ka poora data rakho
# # Left table ka sirf matching lao
# # Match nahi mila → NaN

# | id | name  |
# | -- | ----- |
# | 1  | Ram   |
# | 2  | Shyam |
# | 3  | Mohan |


# | id | marks |
# | -- | ----- |
# | 1  | 80    |
# | 3  | 90    |
# | 4  | 70    |


# | id | name  | marks |
# | -- | ----- | ----- |
# | 1  | Ram   | 80    |
# | 3  | Mohan | 90    |
# | 4  | NaN   | 70    |
```

```python
pd.merge(df1,df2, how = "outer")
# Outer Join kya hota hai?
# Left + Right dono ka poora data
# Match ho ya na ho — sab aayega
# Jahan match nahi → NaN
# Outer Join me DONO tables ka data aata hai
# ✔ match ho ya
# ✔ match na ho

# 👉 Jahan match nahi hota, wahan NULL aa jata hai.

# | student_id | name  |
# | ---------- | ----- |
# | 1          | Rahul |
# | 2          | Sonu  |
# | 3          | Amit  |



# | student_id | marks |
# | ---------- | ----- |
# | 1          | 85    |
# | 3          | 90    |
# | 4          | 70    |


# | student_id | name  | marks |
# | ---------- | ----- | ----- |
# | 1          | Rahul | 85    |
# | 2          | Sonu  | NULL  |
# | 3          | Amit  | 90    |
# | 4          | NULL  | 70    |
```

```python
pd.merge(df1, df2, how = "left",left_on = "key2",right_on = "key4")
# Step 1️⃣ pd.merge ka matlab

# 👉 Do DataFrame ko jodna (join karna)
# Bilkul SQL join jaisa.

# Step 2️⃣ how = "left" (sabse important)

# 👉 LEFT JOIN

# 🧠 Rule:

# df1 ka poora data aayega

# df2 se sirf matching rows aayengi

# Match nahi mila → NaN

# Step 3️⃣ left_on = "key2"

# 👉 df1 ka key2 column use hoga match karne ke liye

# Matlab:


# | key2 | name  |
# | ---- | ----- |
# | 1    | Ram   |
# | 2    | Shyam |
# | 3    | Mohan |

# | key4 | marks |
# | ---- | ----- |
# | 1    | 80    |
# | 3    | 90    |


# | key2 | name  | key4 | marks |
# | ---- | ----- | ---- | ----- |
# | 1    | Ram   | 1    | 80    |
# | 2    | Shyam | NaN  | NaN   |
# | 3    | Mohan | 3    | 90    |
```
